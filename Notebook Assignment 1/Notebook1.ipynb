```
####1.3####
# define a function to check if a column that should be binary
# contains only binary values
def checkIfBinary(s):
  for i in range(len(s)):
    if(s[i] < 0 or s[i] > 1): 
      s[i] = float("NaN")

# any values in age series <= 0 converted to NaN
for i in range(len(heart_df.age)):
  if(heart_df.age[i] <= 0): 
    heart_df.age[i] = float("NaN")

# any non-binary values in anaemia series converted to NaN
checkIfBinary(heart_df.anaemia)

# any values in creatinine_phosphokinase <= 0 converted to NaN
for i in range(len(heart_df.creatinine_phosphokinase)):
  if(heart_df.creatinine_phosphokinase[i] <= 0): 
    heart_df.creatinine_phosphokinase[i] = float("NaN")

# any non-binary values in diabetes series converted to NaN
checkIfBinary(heart_df.diabetes)

# any values in 100 < ejection_fraction < 0 converted to NaN
for i in range(len(heart_df.ejection_fraction)):
  if(heart_df.ejection_fraction[i] < 0 or heart_df.ejection_fraction[i] > 100): 
    heart_df.ejection_fraction[i] = float("NaN")

# any non-binary values in diabetes series converted to NaN
checkIfBinary(heart_df.high_blood_pressure)  

# any values in platelets <= 0 converted to NaN
for i in range(len(heart_df.platelets)):
  if(heart_df.platelets[i] <= 0): 
    heart_df.platelets[i] = float("NaN")

# any values in serum_creatinine <= 0 converted to NaN
for i in range(len(heart_df.serum_creatinine)):
  if(heart_df.serum_creatinine[i] <= 0): 
    heart_df.serum_creatinine[i] = float("NaN")

# any values in serum_sodium <= 0 converted to NaN
for i in range(len(heart_df.serum_sodium)):
  if(heart_df.serum_sodium[i] <= 0): 
    heart_df.serum_sodium[i] = float("NaN")

# any non-binary values in sex series converted to NaN
checkIfBinary(heart_df.sex) 

# any non-binary values in smoking series converted to NaN
checkIfBinary(heart_df.smoking) 

# any values in time <= 0 converted to NaN
for i in range(len(heart_df.time)):
  if(heart_df.time[i] <= 0): 
    heart_df.time[i] = float("NaN")

# any non-binary values in DEATH_EVENT series converted to NaN
checkIfBinary(heart_df.DEATH_EVENT) 

heart_df.to_csv('heart_failure_dataset_NaNs.csv', index=False)
```

```
####1.4####
# convert columns to store floats, needed to read zero mean and unit variance
x = heart_df.astype('float64')

# store mean and standard deviation of each series before iterating through series
mean = x.age.mean()
std = x.age.std()

# iterate through series and normalize values
for i in range(len(x.age)):
  if x.age[i] != float("NaN"): 
    x.age[i] = (float(x.age[i]) - mean) / std

mean = x.creatinine_phosphokinase.mean()
std = x.creatinine_phosphokinase.std()
for i in range(len(x.creatinine_phosphokinase)):
  if x.creatinine_phosphokinase[i] != float("NaN"): 
    x.creatinine_phosphokinase[i] = (float(x.creatinine_phosphokinase[i]) - mean) / std

mean = x.ejection_fraction.mean()
std = x.ejection_fraction.std()
for i in range(len(x.ejection_fraction)):
  if x.ejection_fraction[i] != float("NaN"): 
    x.ejection_fraction[i] = (float(x.ejection_fraction[i]) - mean) / std

mean = x.platelets.mean()
std = x.platelets.std()
for i in range(len(x.platelets)):
  if x.platelets[i] != float("NaN"): 
    x.platelets[i] = (float(x.platelets[i]) - mean) / std

mean = x.serum_creatinine.mean()
std = x.serum_creatinine.std()
for i in range(len(x.serum_creatinine)):
  if x.serum_creatinine[i] != float("NaN"): 
    x.serum_creatinine[i] = (float(x.serum_creatinine[i]) - mean) / std

mean = x.serum_sodium.mean()
std = x.serum_sodium.std()
for i in range(len(x.serum_sodium)):
  if x.serum_sodium[i] != float("NaN"): 
    x.serum_sodium[i] = (float(x.serum_sodium[i]) - mean) / std

mean = x.time.mean()
std = x.time.std()
for i in range(len(x.time)):
  if x.time[i] != float("NaN"): 
    x.time[i] = (float(x.time[i]) - mean) / std

print(x.head(20))
  
```

```
####1.5####
# pop death event column from dataframe
death_event = x.pop('DEATH_EVENT')
# create zeroed len(x.columns) x 2 array of type int64
death = np.zeros(((x.last_valid_index() + 1), 2), dtype='int64')
# iterate and one-hot-encode values of death_event into death array
for i in range(len(death_event)):
  if death_event[i] == 0:
    death[i][0] = 1
    death[i][1] = 0
  else:
    death[i][0] = 0
    death[i][1] = 1
# insert death columns into dataframe
x.insert(loc = len(x.columns), column='NO_DEATH', value=death[:, 0])
x.insert(loc = len(x.columns), column='DEATH', value=death[:, 1])
# save csv file
x.to_csv('heart_failure_dataset_processed.csv', index=False)
```
